Text Scraping - Web Crawling

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

products = []

for page in range(1, 4):
    url = requests.get(f'https://skintific.com/collections/hot-sale?page={page}')
    bs = BeautifulSoup(url.text, 'html.parser')

    for product in bs.find_all(class_='product-item-meta'):

        name = product.find(class_='product-item-meta__title').text

        price_list = product.find(class_='price-list__center')
        price = price_list.find(class_='price').text.strip().replace('Sale price', '').strip().replace('From', '').strip() if price_list else "n/a"

        price_compare = price_list.find(class_='price--compare')
        regular_price = price_compare.text.strip().replace('Regular price', '').strip() if price_compare else price

        rating_stars = product.find('div', class_='rating__stars')
        if rating_stars:
            rating = rating_stars.get('aria-label')
        else:
            rating = 'n/a'

        products.append({
            'Name': name,
            'Price': price,
            'Pre-discount price': regular_price,
            'Rating': rating
        })

df = pd.DataFrame(products)
df.head(58)
```
